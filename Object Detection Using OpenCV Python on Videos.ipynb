```python
import cv2 
import matplotlib.pyplot as plt
```

```python
config_file = 'ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
frozen_model = 'frozen_inference_graph.pb'
```

```python
model = cv2.dnn_DetectionModel(frozen_model,config_file)
```

```python
classLabels = []
file_name = 'labels.txt'
with open(file_name,'rt') as fpt:
    classLabels= fpt.read().rstrip('\n').split('\n')
```

```python
print(classLabels)
```

```python
print(len(classLabels))
```

```python
model.setInputSize(320,320)
model.setInputScale(1.0/127.5)
model.setInputMean((127.5,127.5,127.5))
model.setInputSwapRB(True)
```

```python
import cv2

cap = cv2.VideoCapture('vid1.mp4')
if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError('Can\'t open the video')

font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN

while True:
    ret, frame = cap.read()
    
    ClassIndex, confidence, bbox = model.detect(frame, confThreshold=0.55)
    
    print(ClassIndex)
    
    if len(ClassIndex) != 0:
        for i, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox):
            if i <= 80:  # Assuming there are 80 classes
                cv2.rectangle(frame, boxes, (255, 0, 0), 2)
                cv2.putText(frame, classLabels[i-1], (boxes[0] + 10, boxes[1] + 40), font, fontScale=font_scale, color=(0, 255, 0), thickness=3)

    cv2.imshow('Object Detection by Me', frame)
    
    if cv2.waitKey(2) & 0xff == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
